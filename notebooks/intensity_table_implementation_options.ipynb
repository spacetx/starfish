```python
def append_multiindex_level(multiindex, data, name):
    """stupid thing necessary because pandas doesn't do this"""
    frame = multiindex.to_frame()
    frame[name] = data
    frame.set_index(name, append=True, inplace=True)
    return frame.index
```
